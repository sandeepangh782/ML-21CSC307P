'gdp-for-year' is a categorical column instead of int column. We convert this column into int dtype

```python
suicide[' gdp_for_year ($) ']=suicide[' gdp_for_year ($) '].apply(lambda val: val.replace(',', ''))
suicide[' gdp_for_year ($) '] = pd.to_numeric(suicide[' gdp_for_year ($) '])
suicide[' gdp_for_year ($) ']
```

Converting the categorical values to numeric data

```python

# Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
```

```python
suicide['country_encoded']= label_encoder.fit_transform(df['country'])
suicide['age_encoded']= label_encoder.fit_transform(df['age'])
suicide['generation_encoded']= label_encoder.fit_transform(df['generation'])
suicide.head()
```
